```python
import pandas as pds
import numpy as np
```

```python
df = pds.read_csv("data/export.sql/BIOSAMPLE_DATA_TABLE.dsv", sep="\t")
```

```python
df.head()
```

```python
## find number of records
len(df)
```

```python
## exclude columns that we don't care about
unwanted = ['BIOSAMPLE_ID', 'BIOSAMPLE_NAME', 'ADD_DATE', 'MOD_DATE', 'DESCRIPTION', 'ECOSYSTEM', 
            'ECOSYSTEM_CATEGORY', 'ECOSYSTEM_TYPE', 'ECOSYSTEM_SUBTYPE', 'SPECIFIC_ECOSYSTEM', 
            'PUBLIC_SP_COUNT', 'IS_TEST', 'ADMIN_DAP_COUNT', 'PUBLIC_AP_COUNT', 'ADMIN_SP_COUNT', 'MOD_BY',
            'ACTIVE', 'GOLD_ID', 'IS_PUBLIC', 'SUBMITTER_ID', 'PUBLIC_DAP_COUNT', 'ADMIN_AP_COUNT', 'SPECIMEN']
cols = [c for c in df.columns if c not in unwanted]
```

```python
# cols
subdf = df[cols]
```

```python
## get null counts
series = subdf.isnull().sum().sort_values()
countdf = pds.DataFrame({'field': series.index, 'null_count': series.values})
```

```python
## get counts of non-nulls
countdf['non_null_count'] = len(df) - countdf.null_count
```

```python
## get perecentage of nulls
countdf['percent_null'] = (countdf['null_count'] / (countdf['null_count'] + countdf['non_null_count'])) * 100
countdf['percent_null'] = \
    countdf['percent_null'].map(lambda x: "0%" if np.isinf(x) else "{0:.2f}%".format(x)) # format as percent
```

```python
pds.set_option('display.max_rows', None)
countdf.sort_values(by=['non_null_count'], ascending=False)
```
